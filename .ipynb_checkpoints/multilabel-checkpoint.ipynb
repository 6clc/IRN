```python
from fastai.vision import *
```

```python
# planet = untar_data(URLs.PLANET_TINY)
```

```python
data_dir = "D:\\workspace\\dataset\\adas_parsing_data_0409_to0416\\voc_fmt"
```

```python
dl_tfms = get_transforms( max_lighting=0.1, max_zoom=1.05, max_warp=0.)
```

```python
data = (ImageList.from_csv(data_dir, 'label.csv', folder='JPEGImages', suffix='.jpg')
        .split_by_rand_pct()
        .label_from_df(label_delim=' ')
        .transform(dl_tfms, size=512)
        .databunch(bs=16) # todo change to 8
        .normalize(imagenet_stats))
```

```python
# planet := WindowsPath('C:/Users/liuchao/.fastai/data/planet_tiny')
```

```python
data.show_batch(rows=2, figsize=(9, 7))
```

```python
learn = cnn_learner(data, models.resnet50, metrics=MultiLabelFbeta(beta=2))
```

```python
learn.fit(5)
```

```python
for name, para in learn.model.named_parameters():
    if para.requires_grad:
        print(name)
```

```python
for name, para in learn.model.named_parameters():
    if not para.requires_grad:
        print(name)
```

```python
learn.freeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5,1e-1)
learn.save('mini_train')
```

```python
learn.unfreeze()
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, 1e-2)
```

```python
learn.save('stage2')
```

```python
learn.freeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```
